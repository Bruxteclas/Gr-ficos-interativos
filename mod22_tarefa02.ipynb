Vamos comparar os gráficos de desvio padrão em janela móvel com os gráficos de média móvel para cada ativo. Vamos usar o Plotly para criar um único gráfico que mostra ambos os conjuntos de dados.

````python
# Carregar dados dos arquivos CSV
apple = pd.read_csv(os.path.join(diretorio_destino, 'apple_data.csv'), index_col='Date', parse_dates=True)
google = pd.read_csv(os.path.join(diretorio_destino, 'google_data.csv'), index_col='Date', parse_dates=True)
coca = pd.read_csv(os.path.join(diretorio_destino, 'coca_data.csv'), index_col='Date', parse_dates=True)

# Calcular a média móvel com janela móvel de 90 dias
apple['rolling_mean'] = apple['Adj Close'].rolling(window=90).mean()
google['rolling_mean'] = google['Adj Close'].rolling(window=90).mean()
coca['rolling_mean'] = coca['Adj Close'].rolling(window=90).mean()

# Calcular o desvio padrão com janela móvel de 90 dias
apple['rolling_std'] = apple['Adj Close'].rolling(window=90).std()
google['rolling_std'] = google['Adj Close'].rolling(window=90).std()
coca['rolling_std'] = coca['Adj Close'].rolling(window=90).std()

# Criar gráfico com média móvel e desvio padrão
fig = go.Figure()

# Adicionar média móvel ao gráfico
fig.add_trace(go.Scatter(x=apple.index, y=apple['rolling_mean'], mode='lines', name='AAPL - Média Móvel'))
fig.add_trace(go.Scatter(x=google.index, y=google['rolling_mean'], mode='lines', name='GOOGL - Média Móvel'))
fig.add_trace(go.Scatter(x=coca.index, y=coca['rolling_mean'], mode='lines', name='KO - Média Móvel'))

# Adicionar desvio padrão ao gráfico
fig.add_trace(go.Scatter(x=apple.index, y=apple['rolling_std'], mode='lines', name='AAPL - Desvio Padrão'))
fig.add_trace(go.Scatter(x=google.index, y=google['rolling_std'], mode='lines', name='GOOGL - Desvio Padrão'))
fig.add_trace(go.Scatter(x=coca.index, y=coca['rolling_std'], mode='lines', name='KO - Desvio Padrão'))

# Adicionar períodos de maior volatilidade ao gráfico (substitua 'periods_of_high_volatility' pelos períodos desejados)
periods_of_high_volatility = ['2020-05-11', '2020-05-12', '2020-05-13', '2021-06-15']
for period in periods_of_high_volatility:
    fig.add_shape(dict(type='line', x0=period, x1=period, y0=0, y1=1, xref='x', yref='paper', line=dict(color='gray', width=2, dash='dash')))

# Configurar layout
fig.update_layout(title='Comparação de Média Móvel e Desvio Padrão com Períodos de Maior Volatilidade',
                  xaxis_title='Data',
                  yaxis_title='Valor',
                  template='plotly_white')

# Exibir gráfico
fig.show()
````

````python
# Função para calcular a média móvel e o desvio padrão
def calcular_medias_desvios(df, janela):
    media_movel = df['Close'].rolling(window=janela).mean()
    desvio_padrao = df['Close'].rolling(window=janela).std()
    return media_movel, desvio_padrao

# Diretório de destino
diretorio_destino = 'C:\\Users\\User\\OneDrive\\Jupyter\\Modulo 22\\'

# Carregar dados do arquivo CSV
apple = pd.read_csv(os.path.join(diretorio_destino, 'apple_data.csv'), index_col='Date', parse_dates=True)

# Escolher janela para média móvel e desvio padrão (por exemplo, 90 dias)
janela = 90

# Calcular média móvel e desvio padrão
media_movel, desvio_padrao = calcular_medias_desvios(apple, janela)

# Adicionar colunas ao DataFrame
apple['MediaMovel'] = media_movel
apple['DesvioPadrao'] = desvio_padrao

# Identificar períodos de maior volatilidade (por exemplo, desvio padrão acima de um limiar)
limiar_volatilidade = 5  # Ajuste conforme necessário
periodos_volatilidade = apple[apple['DesvioPadrao'] > limiar_volatilidade].index

# Exibir as datas dos períodos de maior volatilidade
print("Períodos de Maior Volatilidade:")
print(periodos_volatilidade)
````
